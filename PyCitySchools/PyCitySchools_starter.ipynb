```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
school_data_complete_df["school_name"].unique()
school_list_df = pd.DataFrame(school_data_complete_df["school_name"].unique())
total_Schools = len(school_list_df)

# Calculate the total number of students
school_data_complete_df["Student ID"].unique()
total_Students_df = pd.DataFrame(school_data_complete_df["Student ID"].unique())
total_Students =len(total_Students_df)

# Calculate the total budget
total_budget = school_data["budget"].sum()

# Calculate the average maths score
average_Maths_Score = school_data_complete_df["maths_score"].mean()

# Calculate the average reading score
average_Reading_Score = school_data_complete_df["reading_score"].mean()

# Calculate the percentage of students with a passing maths score (50 or greater)
passing_Maths = len(school_data_complete_df[school_data_complete_df["maths_score"] >= 50])
percentage_Passing_Maths = passing_Maths / total_Students * 100

# Calculate the percentage of students with a passing reading score (50 or greater)
passing_Reading = len(school_data_complete_df[school_data_complete_df["reading_score"] >= 50])
percentage_Passing_Reading = passing_Reading / total_Students * 100

# Calculate the overall passing rate (overall average score), i.e. (avg. maths score + avg. reading score)/2
Overall_student_passing = school_data_complete_df.loc[(school_data_complete_df["reading_score"]>= 50) &
                                                      (school_data_complete_df["maths_score"]>= 50) , :]
Overall_student_passing= Overall_student_passing["student_name"].count()
Overall_student_passing
percentage_Passing_Rate =(Overall_student_passing/total_Students)*100
percentage_Passing_Rate


# Create a dataframe to hold the above results
lga_summary_df = pd.DataFrame({
    "Total Schools": [total_Schools],
    "Total Students": [total_Students],
     "Total Budget": [total_budget],
     "Average Maths Score": [average_Maths_Score],
      "Average Reading Score": [average_Reading_Score],
      "% Passing Maths": [percentage_Passing_Maths],
      "% Passing Reading": [percentage_Passing_Reading],
      "% Overall Passing Rate": [percentage_Passing_Rate]
})

# Cleaner formatting
lga_summary_df["Total Budget"] = lga_summary_df["Total Budget"].map("${:,.2f}".format)

lga_summary_df[["Total Schools",
     "Total Students",
      "Total Budget",
      "Average Maths Score",
      "Average Reading Score",
      "% Passing Maths",
      "% Passing Reading",
      "% Overall Passing Rate"]]
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Creating a table summary of the key metrics of each school
# By school type
school_type = school_data.set_index(["school_name"])["type"]

# Total students
total_students = school_data_complete_df.groupby(["school_name"]).count()["student_name"]

# Total schools budget
total_school_budget= school_data_complete_df.groupby(["school_name"]).mean()["budget"]
# Budget per student
budget_per_student = total_school_budget / total_students

#average Maths score
average_Maths_Score = school_data_complete_df.groupby(["school_name"]).mean()["maths_score"]
# average reading score
average_Reading_Score = school_data_complete_df.groupby(["school_name"]).mean()["reading_score"]

# percentage passing maths
students_passing_Maths = school_data_complete_df[school_data_complete_df["maths_score"] >= 50]
percentage_passing_Maths = students_passing_Maths.groupby(["school_name"]).count()["student_name"]/total_students*100

# percentage passing reading
students_passing_Reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 50]
percentage_Passing_Reading = students_passing_Reading.groupby(["school_name"])["reading_score"].count()/total_students*100

#calculate percentage of overall
Overall_student_passing = school_data_complete_df.loc[(school_data_complete_df["reading_score"]>= 50) &
                                                      (school_data_complete_df["maths_score"]>= 50) , :]
percentage_Passing_Rate = Overall_student_passing.groupby(["school_name"])["reading_score"].count()/total_students*100


# Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Type": school_type,
      "Total Students": total_students,
      "Total School Budget": total_school_budget,
      "Per Student Budget": budget_per_student,
      "Average Maths Score": average_Maths_Score,
      "Average Reading Score": average_Reading_Score,
      "% Passing Maths": percentage_passing_Maths,
      "% Passing Reading": percentage_Passing_Reading,
      "% Overall Passing Rate": percentage_Passing_Rate})
school_summary_df

# Optional: give the displayed data cleaner formatting
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)

school_summary_df[["School Type",
      "Total Students",
      "Total School Budget",
      "Per Student Budget",
      "Average Maths Score",
      "Average Reading Score",
      "% Passing Maths",
      "% Passing Reading",
      "% Overall Passing Rate"]]
```

```python
# Sort and display the top five schools in overall passing rate
school_summary_df = school_summary_df.sort_values(["% Overall Passing Rate"], ascending=False)
school_summary_df.head()
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create data series of scores by grade levels using conditionals
year_9 = school_data_complete_df.loc[(school_data_complete_df["year"] == 9)].groupby(["school_name"]).mean()["maths_score"]
year_10 = school_data_complete_df.loc[(school_data_complete_df["year"] == 10)].groupby(["school_name"]).mean()["maths_score"]
year_11 = school_data_complete_df.loc[(school_data_complete_df["year"] == 11)].groupby(["school_name"]).mean()["maths_score"]
year_12 = school_data_complete_df.loc[(school_data_complete_df["year"] == 12)].groupby(["school_name"]).mean()["maths_score"]



# Combine series into single DataFrame
scores_by_year = pd.DataFrame({"Year 9": year_9, "Year 10": year_10,
                                "Year 11": year_11, "Year 12": year_12})
scores_by_year = scores_by_year[["Year 9","Year 10","Year 11","Year 12"]]
scores_by_year.index.name = None
# Display the DataFrame
scores_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins=[0,585,630,645,680]
group_names = ["<585","585-630","630-645","645-680"]
school_summary_df["scores by school spending"] = pd.cut(budget_per_student,bins, labels=group_names, include_lowest=True)
school_summary_df
```

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student)
avg_maths_score_by_spending = school_summary_df.groupby(["scores by school spending"]).mean()['Average Maths Score']
avg_reading_score_by_spending = school_summary_df.groupby(["scores by school spending"]).mean()['Average Reading Score']
avg_maths_Percent_by_spending = school_summary_df.groupby(["scores by school spending"]).mean()['% Passing Maths']
avg_reading_Percent_by_spending = school_summary_df.groupby(["scores by school spending"]).mean()['% Passing Reading']
avg_overall_Percent_by_spending = school_summary_df.groupby(["scores by school spending"]).mean()['% Overall Passing Rate']

Score_by_school_spending = pd.DataFrame({"Average Maths Score":avg_maths_score_by_spending,
                                         "Average Reading Score":avg_reading_score_by_spending,
                                         "% Passing Maths":avg_maths_Percent_by_spending,
                                         "% Passing Reading":avg_reading_Percent_by_spending,
                                         "% Overall Passing Rate":avg_overall_Percent_by_spending
                                        })

Score_by_school_spending = Score_by_school_spending[["Average Maths Score","Average Reading Score","% Passing Maths",
                                                   "% Passing Reading","% Overall Passing Rate"]]
Score_by_school_spending.index.name=None
Score_by_school_spending
```
